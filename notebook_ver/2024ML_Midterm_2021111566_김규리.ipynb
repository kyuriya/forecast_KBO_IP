### 1. 문제정의 (Problem Define)
투수의 '투구이닝수'는 MLB에서 뿐만 아니라 투수를 평가할 때 주요하게 여겨지는 지표 중 하나임.투수의 능력치가 종합적으로 반영되는 지표이기 때문임. 따라서 투수들의 전년도 성적, 키와 몸무게와 같은 개인적인 요소들 그리고 소속된 팀의 상황에 따라 해당 투수의 다음년도 정규시즌에 등판할 이닝 수를 예측해보려고 함. 

- y = f(x)
- y: 2022년도 투수의 정규시즌 소화 이닝수
- x: 2021년도 투수의 정규시즌 개인 성적, 성적이 아닌 개인적인 요소들(나이,키,몸무게 등), 소속된 팀의 요소들
- f: 회귀모델링(Regression)

### 2.3 데이터 변수 의미 파악
#### 1. 선수 개인 성적 
- 선수명
- 팀명
- 승 : 승리투수로 기록된 경기
- 패 : 패전투수로 기록된 경기
- 세(세이브) : 마무리 투수에게 중요한 기록으로, 8,9회에 등판하여 역전 허용하지 않고 승리 지켜낸 경기수
- 홀드 : 3점차 이하로 팀이 이기고 있어, 세이브가 가능한 상황에서 중간에 올라온 투수가 실점하지 않고 팀이 이기고 있는 상황을 유지한 경기수
- 블론 : 마무리 투수가 세이브 기록할 수 있는 상황에서 동점 또는 역전 허락한 경기수
- 경기 : 참여한 경기수
- 선발 : 선발 투수로서 등판한 경기수
- 이닝 : 투수가 소화한 2021년 정규 시즌의 총 누적 이닝수 (1 경기 = 9이닝, 1이닝 = 3 아웃카운트)
- 삼진/9 : 9이닝 당 삼진 갯수
- 볼넷/9 : 9이닝 당 볼넷 준 갯수    
- 홈런/9 : 9이닝 당 맞은 홈런 갯수   
- BABIP : 타자가 친공이 선 안으로 들어왔을 때, 이것이 안타가 될 확률(투수의 BABIP이 높다면 안타를 많이 맞았다는 의미)
- LOB%  : 잔루 처리율(출루시킨 루상의 주자 중 득점하지 않은 비율로써 얼마나 잔루로 처리했느냐)
- ERA  : 평균 자책점(*자책점 : 실점 중 투수의 직접적인 실력 및 책임으로만 얻은 실점),(9이닝 동안 투구했을 때 평균 몇 점의 자책점을 주는가)   
- WHIP : 평균적으로 한 이닝 동안 얼마나 많은 주자를 베이스에 허용하는지
- FIP : 팀의 수비와 무관한 투수의 평균 자책점 
- WAR  : 대체선수 대비 승리기여도(투수,타자 모두 평가 가능한 종합지표) 
- 타자 : 2021년 정규시즌 동안 상대한 타자 수
- 안타 : 안타를 허용한 개수
- 2루타 : 2루타를 허용한 개수
- 3루타 : 3루타를 허용한 개수
- 이닝2022 : 2022년도 투수의 이닝수

#### 2. 팀 정보
- 정규시즌승률  : 2021년도 정규 시즌 팀의 승률
- team_ERA : 2021년도 각 팀 투수들의 평균 ERA
- 총투수 : 각 팀이 보유한 투수의 수
#### 3. 선수 개인 정보
- 나이 : 2021년 기준 나이       
- 키     
- 몸무게

##### 수집된 데이터 전처리

```python
#1. 2021, 2022 각년도 데이터에서 동명이인 선수명 변경
duplicates2021 = p2021[p2021.duplicated('선수명', keep=False)]
print(duplicates2021)

#10건 가량 존재
#2021년에서 '선수명'+'팀명첫글자'
p2021.loc[25,'선수명'] = '김대우삼'
p2021.loc[26,'선수명'] = '김대우롯'
p2021.loc[71,'선수명'] = '김태훈H'
p2021.loc[72,'선수명'] = '김태훈S'
p2021.loc[105,'선수명'] = '박세웅롯'
p2021.loc[106,'선수명'] = '박세웅삼'
p2021.loc[198,'선수명'] = '이승헌롯'
p2021.loc[199,'선수명'] = '이승헌N'
p2021.loc[294,'선수명'] = '한승혁K'
p2021.loc[295,'선수명'] = '한승혁롯'

duplicates2021 = p2021[p2021.duplicated('선수명', keep=False)]
print(duplicates2021)
```

```python
#2022년도 마찬가지
duplicates2022 = p2022[p2022.duplicated('선수명2022', keep=False)]
print(duplicates2022)

# 6건 존재
#2022년에서 '선수명'+'팀명첫글자'
p2022.loc[19,'선수명2022'] = '김대우삼'
p2022.loc[20,'선수명2022'] = '김대우롯'
p2022.loc[61,'선수명2022'] = '김태훈H'
p2022.loc[62,'선수명2022'] = '김태훈S'
p2022.loc[92,'선수명2022'] = '박세웅롯'
p2022.loc[93,'선수명2022'] = '박세웅삼'

#병합 위해 선수명 컬럼에 할당
p2022['선수명'] = p2022['선수명2022']
duplicates2022 = p2022[p2022.duplicated('선수명', keep=False)]
print(duplicates2022)
```

```python
#2.선수명단 확인
#2021,2022년도의 선수단 명단 차이 보기 위해 병합
p = pd.merge(p2021, p2022,how='outer', on='선수명') 
p
#2021년만 활동한 선수 또는 2021년 이후로는 정규 시즌에서 경기하지 못했거나.. 등등의 사유로 없는 값들 116건
print(p[p['선수명2022'].isnull()])

#2022년의 선수명 없거나, 이닝수 없는 결측치 제거해줌
p= p.dropna(subset=['이닝','선수명2022']).reset_index(drop=True)
p
```

```python
# '팀명2022' 컬럼 제거
# '선수명2022'도 제거
pitch_origin = p.drop(['선수명2022','팀명2022'], axis=1)
pitch_origin
```

```python
#3.2021년 당시 팀 투수들 평균 자책점, 정규시즌 승률 등을 독립변수로 추가
# 2021년도 데이터 가져오기
# kbopitch = kbopitch[kbopitch['year']==2021][['team','average_age','win_loss_percentage','ERA']]
kbopitch = kbopitch[kbopitch['year']==2021][['team','win_loss_percentage','ERA']]
#병합 위해 원본 데이터의 팀명 변경
team_n = ['LG','KT','두산','삼성','NC','Hero','SSG','한화','KIA','롯데']
for a in range(10):
    kbopitch.loc[a,'팀명'] = team_n[a]
print(kbopitch)

#각 팀에 맞게 추가해주기
for z in range(len(pitch_origin)):
    for q in range(len(kbopitch)):
        if pitch_origin.loc[z,'팀명']== kbopitch.loc[q,'팀명']:
            # pitch_origin.loc[z,'투수평균나이'] = kbopitch.loc[q,'average_age']
            pitch_origin.loc[z,'정규시즌승률'] = kbopitch.loc[q,'win_loss_percentage']
            pitch_origin.loc[z,'team_ERA'] = kbopitch.loc[q,'ERA']
print(pitch_origin)
```

```python
for a in range(len(pitch_origin)):
    for z in range(len(team_num)):
        if pitch_origin.loc[a,'팀명']== team_num.loc[z,'팀명']:
            pitch_origin.loc[a,'총투수'] = team_num.loc[z,'총투수']
pitch_origin
```

```python
#사용할  정보만 추출하여 컬럼으로 재생성
#전처리
#선수명 전처리
pinfo['선수명'] = pinfo['선수명'].str.replace(',', '')
pinfo['선수명'] = pinfo['선수명'].str.replace("'", '')
pinfo['선수명'] = pinfo['선수명'].str.replace(" ", '')
pinfo

#선수 정보 쪼개기 - 생년월일 / 키/ 몸무게/ 소속 
#생년월일

for a in range(len(pinfo)):
  #생년월일
    pinfo.loc[a,'태어난년도'] = pinfo.loc[a,'선수정보'].split()[1]
    pinfo.loc[a,'태어난년도'] = pinfo.loc[a,'태어난년도'].split('-')[0]
    #2021년 당시 나이 계산
    pinfo.loc[a,'나이'] = 2021 - int(pinfo.loc[a,'태어난년도']) +1

  #키
    pinfo.loc[a,'키/몸무게'] = pinfo.loc[a,'선수정보'].split()[9]
    pinfo.loc[a,'키'] = pinfo.loc[a,'키/몸무게'][:3]
    if pinfo.loc[a,'키/몸무게'][6] == '1':
        pinfo.loc[a,'몸무게'] = pinfo.loc[a,'키/몸무게'][6:9]
    else:
        pinfo.loc[a,'몸무게'] = pinfo.loc[a,'키/몸무게'][6:8]

  
pinfo
```

### [다중공산성 1] VIF: Variance Inflation Factor
VIF>10인 독립변수들을 제거해줌. 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['features'] = X_train.columns
vif["VIF Factor"] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif.round(1)  
```

```python
#다중공선성 큰 순으로 하나씩 차례대로 제거
numerical_columns =['승', '패', '세', '홀드', '블론', '경기',   '삼진/9', '볼넷/9', '홈런/9',  'LOB%',
                  'ERA','WHIP','WAR','2루타','3루타','정규시즌승률','나이','키','몸무게','team_ERA','총투수']

X = pitch_df[numerical_columns]
y = pitch_df['이닝2022']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

vif = pd.DataFrame()
vif['features'] = X_train.columns
vif["VIF Factor"] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif.round(1)  
```
